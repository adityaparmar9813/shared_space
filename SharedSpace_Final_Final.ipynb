## Elbow Method and Model Implementation

```python
from sklearn.cluster import KMeans
```

```python
cost_female = []

for i in range(1, 51):
    KM = KMeans(n_clusters = i, max_iter = 500)
    KM.fit(female_df)
    
    cost_female.append(KM.inertia_)
    
plt.plot(range(1, 51), cost_female, color ='g', linewidth ='3')
plt.xlabel("Value of K")
plt.ylabel("Squared Error (Cost)")
plt.show()
```

```python
cost_male = []

for i in range(1, 51):
    KM = KMeans(n_clusters = i, max_iter = 500)
    KM.fit(male_df)
    
    cost_male.append(KM.inertia_)
    
plt.plot(range(1, 51), cost_male, color ='g', linewidth ='3')
plt.xlabel("Value of K")
plt.ylabel("Squared Error (Cost)")
plt.show()
```

```python
kmeans_female = KMeans(n_clusters=30)
kmeans_male = KMeans(n_clusters=25)
```

```python
kmeans_female.fit(female_df)
```

```python
kmeans_male.fit(male_df)
```
